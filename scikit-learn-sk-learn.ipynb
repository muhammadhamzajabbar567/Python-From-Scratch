**Applying KNN - Algorithm**

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

x_test,x_train,y_test,y_train = train_test_split(x,y, test_size=0.45,random_state=1)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
 

knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(x_train,y_train)
y_pred = knn_classifier.predict(x_test)
print('Accuracy: ',metrics.accuracy_score(y_test,y_pred))

sample =[[5,5,3,8], [2,8,6,4]]
preds = knn_classifier.predict(sample)  
pred_species = [iris.target_names[p] for p in preds] 
print("Predictions:",pred_species)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.45,random_state=1)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn_classi = KNeighborsClassifier(n_neighbors=2)
knn_classi.fit(x_test,y_test)
y_pred = knn_classi.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_pred,y_test))

sample = [[1,2,3,4],[20,1,3,4]]
preds = knn_classi.predict(sample)
predicts = [iris.target_names[p] for p in preds]
print('Predictions:', predicts)
```

**Preprocessing the Data**
**Binarisation **

```python
from sklearn import preprocessing
import numpy as np

arr =  np.array([[2.1, -1.9, 5.5], 
     [-1.5, 2.4, 3.5], 
     [0.5, -7.9, 5.6], 
     [5.9, 2.3, -5.8]])
data_binarized = preprocessing.Binarizer(threshold=0.7).transform(arr)
print('\n Data Binarized \n', data_binarized)
```

```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[0,1.5,-4],[-396,1.69,-1],[2,-5,9],[2,5,-9],[1,2,5]])
data_binarized = preprocessing.Binarizer(threshold=1).transform(arr)
print("\n Data Binarized \n", data_binarized)
```

****MEAN REMOVAL****


```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[1,2,-5],[2,8,8],[-6,-9,-2],[2,-6,7],[0,7,-6]])
print("MEAN = ", arr.mean(axis=0))
print("STANDARD DEVIATION = ",arr.std(axis=0))

# MEAN REMOVAL
data_scaled = preprocessing.scale(arr)
print('MEAN REMOVAL = ', data_scaled.mean(axis=0))
print("STANDARD DEVIATION REMOVEL = ",data_scaled.std(axis=0))
```

*****SCALING******

```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[1,2,3],[4,0,6],[9,-6.,-7],[-1,25,1]])
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,2))
min_max_scaled = min_max_scaler.fit_transform(arr)
print("Min Max Scaled = ", min_max_scaled)
```

```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[1,0,0],[-1,6,-9],[-6,-8,-3],[3,-4,6]])
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
min_max_scaled = min_max_scaler.fit_transform(arr)
print("Min Max Scaled = ",min_max_scaled)
```

**Normalisation**

**L1 Normalisation**

```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[-1,-2,-3],[0,1,2],[-6,8,9],[1,3,2]])

l1_normalize = preprocessing.normalize(arr,norm='l1')
print("\nL1 Normalisation\n",l1_normalize)
```

**L2 Normalisation**

```python
import numpy as np
from sklearn import preprocessing

arr = np.array([[0,1,6],[6,5,8],[2,3,8],[-1,-7,-4],[-1,-3.6,-5]])
l2_normalize = preprocessing.normalize(arr, norm='l2')
print("\nL2 Normalisation\n",l2_normalize)
```

**L1 Normalisation Vs L2 Normalisation**

**Least Absolute Deviations (L1) Vs Least Squares(L2)**

```python

import numpy as np
from sklearn import preprocessing

arr = np.array([[0,1,6],[6,5,8],[2,3,8],[-1,-7,-4],[-1,-3.6,-5]])
l1_normalize = preprocessing.normalize(arr, norm='l1')
l2_normalize = preprocessing.normalize(arr, norm='l2')
print("\nL1 Normalisation\n",l1_normalize)
print("\nL2 Normalisation\n",l2_normalize)
```
